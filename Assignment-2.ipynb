# create calculator that can apply this operation in two numbers using opp  

1. add 
2. subtraction 
3. multiplication 
4. division 
5. check even and odd 
6. factorial 
7. fibonacii 

```python
class calc:
    def __init__(self,x,y=None):
        print('welcome to oop calc')
        self.x=x
        self.y=y
    def add(self):
        print(f'addition : {self.x} + {self.y} = {self.x+self.y}')
    def sub(self):
        print(f'subtraction : {self.x} - {self.y} = {self.x-self.y}')
    def mult(self):
        print(f'multiplication : {self.x} * {self.y} = {self.x*self.y}')
    def div(self):
        print(f'division : {self.x} / {self.y} = {self.x/self.y}')
    def check_even_or_odd(self):
        if self.x%2==0:
            print(f'{self.x} is even')
        else:
            print(f'{self.x} is odd')
    def fact(self):
        fact=1
        for i in range(1,self.x+1):
            fact=fact*i
        print(f'factorial of {self.x} = {fact}')
    
    def fib(self):
        count=0
        num1=0
        num2=1
        fibonaci_seq=[]
        while count<self.x:

            fibonaci_seq.append(num1)
            new_num=num1+num2
            num1=num2
            num2=new_num      
            count+=1
        print(f'fibonaci sequence is : {fibonaci_seq}')





            
        
        
```

```python
obj=calc(9,3)
obj.add()
print('\n')
obj.sub()
print('\n')
obj.mult()
print('\n')
obj.div()
print('\n')
obj=calc(10)
obj.check_even_or_odd()
print('\n')
obj.fact()
print('\n')
obj.fib()

```

# create calculator that can apply this operation in two numbers or more using opp  

1. add 
2. subtraction 
3. multiplication 
4. division 
5. check even and odd 
6. factorial 
7. fibonacii 

```python
class calc2:
    def __init__(self,*x):
        print('welcome to oop calc')
        self.x=x
    def add(self):
        print(f'addition :{sum(self.x)}')
    def sub(self):
        result=self.x[0]
        for i in self.x:
            if self.x.index(i)!=0:
                result-=i
        print(f'subtraction:{result}')
    
    def mult(self):
        result=1
        for i in self.x:
            result*=i
        print(f'Multiplication :{result}')
        
    def div(self):
        result=1
        for i in self.x:
            result/=i
        print(f'division:{result}')
    
    def check_even_or_odd(self):
        for item in self.x:
            if item%2==0:
                print(f'{item} is even')
            else:
                print(f'{item} is odd')
    def fact(self):
        
        for item in self.x:
            fact=1
            for i in range(1,item+1):
                fact=fact*i
            print(f'factorial of {item} = {fact}')
    
    def fib(self):
        for item in self.x:
            count=0
            num1=0
            num2=1
            fibonaci_seq=[]
            while count<item:

                fibonaci_seq.append(num1)
                new_num=num1+num2
                num1=num2
                num2=new_num      
                count+=1
            print(f'fibonaci sequence of {item} is : {fibonaci_seq}')

        
        
        

    
    
    
```

```python
obj=calc2(1,2,3,4)
obj.add()
print('\n')
obj.sub()
print('\n')
obj.mult()
print('\n')
obj.div()
print('\n')
obj=calc2(5,10,15)
obj.check_even_or_odd()
print('\n')
obj.fact()
print('\n')
obj.fib()
```

# Task 3: Rectangle class: 
1. Write a Rectangle class in Python language, allowing you to build a rectangle 
with length and width attributes.
2. Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to 
calculate the area of the rectangle.
3. Create a method display() that display the length, width, perimeter and area of an object created 
using an instantiation on rectangle class.
4. Create a Parallelepipede child class inheriting from the Rectangle class and with 
a height attribute and another Volume() method to calculate the volume of the Parallelepiped

```python
class rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
        print('welcome to Rectangle OOP')
    def perimeter(self):
        global perim 
        perim=2*(self.length+self.width)
        print(f'The Perimeter is 2*({self.length}+{self.width})= {perim}')
    def area(self):
        global ar
        ar=self.length*self.width
        print(f'The Area is {self.length}*{self.width} = {ar}')
    def display(self):
        print('display usage')
        print(f'the length: {self.length}')
        print(f'the width: {self.width}')
        print(f'the perimeter: {perim}')
        print(f'the area: {ar}')
        
        
    
```

```python
obj=rectangle(2,4)
print('\n')
obj.perimeter()
print('\n')
obj.area()
print('\n')
obj.display()
```

```python
class cuboid(rectangle):
    def __init__(self,length,width,height):
        super().__init__(length,width)
        self.height=height
    def vol(self):
        print(f'volume is : {self.length*self.width*self.height}')
        
        
```
